
 # PASO 4: Reducción de dimensionalidad usando PCA (Análisis de Componentes Principales)

```python
#  Estandarizamos las variables (var_), excluyendo ID_code y target

from sklearn.preprocessing import StandardScaler

# Variables de entrada
features = [col for col in df.columns if col.startswith('var_')]

# Escalador
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])
```

```python
# Aplicamos PSA

from sklearn.decomposition import PCA

# Aplicamos PCA con todos los componentes
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
#Visualizo la varianza explicada (para elegir cuántos componentes conservar)

import matplotlib.pyplot as plt
import numpy as np

# Varianza acumulada
varianza_acumulada = np.cumsum(pca.explained_variance_ratio_)

# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(varianza_acumulada, marker='o')
plt.axhline(y=0.95, color='r', linestyle='--')
plt.xlabel('Número de componentes principales')
plt.ylabel('Varianza acumulada')
plt.title('Varianza explicada acumulada por PCA')
plt.grid(True)
plt.show()
```

# PASO 5: Aplicación final de PCA y preparación del dataset reducido

```python
# Seleccionamos solo las columnas con variables tipo 'var_'
features = [col for col in df.columns if col.startswith('var_')]

# Definimos X (características) e y (target)
X = df[features]
y = df['target']
```

```python
# Divido en entrenamiento y validación

from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
# Estandarizar (escalar) los datos

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Ajustamos solo con los datos de entrenamiento
X_train_scaled = scaler.fit_transform(X_train)

# Transformamos los datos de validación usando el mismo escalador
X_valid_scaled = scaler.transform(X_valid)
```

```python
# Aplicar PCA con 3 componentes

from sklearn.decomposition import PCA

# Aplicamos PCA con 3 componentes
pca = PCA(n_components=3)

# Ajustamos solo con los datos escalados de entrenamiento
X_train_pca = pca.fit_transform(X_train_scaled)

# Transformamos la validación con el mismo PCA ya entrenado
X_valid_pca = pca.transform(X_valid_scaled)
```

```python
# Crear nuevos DataFrames para regresión logística

import pandas as pd

columns_pca = [f'pca_{i+1}' for i in range(3)]

# Crear DataFrames
X_train_logistic = pd.DataFrame(X_train_pca, columns=columns_pca)
X_valid_logistic = pd.DataFrame(X_valid_pca, columns=columns_pca)
```

```python
# Verificar la forma de los datos

print("X_train_logistic shape:", X_train_logistic.shape)
print("X_valid_logistic shape:", X_valid_logistic.shape)
print("y_train shape:", y_train.shape)
print("y_valid shape:", y_valid.shape)
```

# Paso 7: Modelo con LightGBM (sin PCA)

```python
# Juntar X e y en un solo dataframe para train
df_train = X_train.copy()
df_train['target'] = y_train

# Juntar X e y en un solo dataframe para valid
df_valid = X_valid.copy()
df_valid['target'] = y_valid

```

```python
# Selección de variables predictoras (sin ID_code ni target)
features = [col for col in df_train.columns if col not in ['ID_code', 'target']]

X_train_lgbm = df_train[features]
X_valid_lgbm = df_valid[features]

y_train_lgbm = df_train['target']
y_valid_lgbm = df_valid['target']
```

```python
# Aquí voy a recrear esos dataframes combinando X_train y y_train, y lo mismo con los de validación

import lightgbm as lgb

# Creamos datasets para LightGBM
train_set = lgb.Dataset(X_train_lgbm, label=y_train_lgbm)
valid_set = lgb.Dataset(X_valid_lgbm, label=y_valid_lgbm)

# Parámetros básicos
params = {
    'objective': 'binary',
    'metric': 'auc',
    'verbosity': -1,
    'boosting_type': 'gbdt',
    'learning_rate': 0.05,
    'num_leaves': 31,
    'class_weight': 'balanced',  # muy importante por el desbalance
    'random_state': 42
}

# Entrenamiento
model_lgb = lgb.train(
    params,
    train_set,
    valid_sets=[train_set, valid_set],
    num_boost_round=500,
    callbacks=[
        lgb.early_stopping(stopping_rounds=30),
        lgb.log_evaluation(period=50)
    ]
)
```

```python
#  ¿Qué significan estos valores?

    #- AUC de entrenamiento (0.558): apenas mejor que lanzar una moneda (0.50).
    #- AUC de validación (0.504): el modelo no está aprendiendo a distinguir bien entre las clases (transacción vs no transacción).
    #- Early stopping en la iteración 23: se detuvo pronto porque el rendimiento no mejoraba significativamente.

# Esto nos dice que el modelo actual no está captando patrones útiles .vv.
```

```python
# Predicciones de probabilidad
y_pred_lgb = model_lgb.predict(X_valid_lgbm)
# Convertimos a clases (0 o 1) con un umbral de 0.5
y_pred_class = (y_pred_lgb >= 0.5).astype(int)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

print("Classification Report:")
print(classification_report(y_valid_lgbm, y_pred_class))

auc_lgb = roc_auc_score(y_valid_lgbm, y_pred_lgb)
print(f"\nROC AUC Score (LightGBM): {auc_lgb:.4f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, roc_curve

# Matriz de Confusión
cm = confusion_matrix(y_valid_lgbm, y_pred_class)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Matriz de Confusión - LightGBM")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.show()

# Curva ROC
fpr, tpr, _ = roc_curve(y_valid_lgbm, y_pred_lgb)
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f'AUC = {auc_lgb:.4f}')
plt.plot([0, 1], [0, 1], 'k--')
plt.title("Curva ROC - LightGBM")
plt.xlabel("Falsos Positivos")
plt.ylabel("Verdaderos Positivos")
plt.legend()
plt.grid()
plt.show()
```

```python
# Veo que mi modelo de LightGBM predice solo ceros (no-transacción). Eso lo confirma:

   # - Matriz de confusión con 0 en la diagonal de clase 1.
   # - Curva ROC plana (AUC ~0.50).

# Asi que me tocará Mejorar LightGBM para que aprenda a detectar la clase 1 aplicando estrategias como SMOTE (oversampling) y Ajustar el umbral de decisión
```

```python
from imblearn.over_sampling import SMOTE

# Aplicamos SMOTE
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train_lgbm, y_train_lgbm)

# Verificamos proporciones
import numpy as np
print("Distribución después de SMOTE:", np.bincount(y_train_res))
```

```python
train_set = lgb.Dataset(X_train_res, label=y_train_res)
valid_set = lgb.Dataset(X_valid_lgbm, label=y_valid_lgbm)

model_lgb_smote = lgb.train(
    params,
    train_set,
    valid_sets=[train_set, valid_set],
    num_boost_round=500,
    callbacks=[
        lgb.early_stopping(stopping_rounds=30),
        lgb.log_evaluation(period=50)
    ]
)
```

```python
y_pred_proba = model_lgb_smote.predict(X_valid_lgbm)
y_pred_class = (y_pred_proba >= 0.3).astype(int)

from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
print(classification_report(y_valid_lgbm, y_pred_class))
print("ROC AUC Score:", roc_auc_score(y_valid_lgbm, y_pred_proba))
```

```python
# Conclusión

# Aunque he aplicado SMOTE y bajado el umbral de decisión, veo que el ROC AUC sigue muy bajo (0.51). 
# Eso nos dice que, aunque el modelo ahora sí predice ambas clases, no está encontrando patrones útiles para diferenciar clientes 
# que hacen una transacción de los que no la hacen.

# Que intentaré ahora para mejorar los resultados del modelo?

# 1. Reducir ruido con Selección de Variables - Muchos de los ~200 var_ son ruido o duplicados en cuanto a información. Si elimino variables poco útiles, el modelo ve mejor las señales.
# 2. Ajustar Hiperparámetros de LightGBM 
# 3. Cambiar estrategia: usar los componentes PCA o crear nuevas variables
```

```python
import pandas as pd
import numpy as np

# Entrena un modelo rápido con los datos resampleados
model_temp = lgb.LGBMClassifier()
model_temp.fit(X_train_res, y_train_res)

# Ordenar variables por importancia
feature_importances = pd.Series(model_temp.feature_importances_, index=X_train_lgbm.columns)
top_features = feature_importances.sort_values(ascending=False).head(40).index.tolist()

# Crear nuevos datasets con solo estas variables
X_train_reduced = X_train_lgbm[top_features]
X_valid_reduced = X_valid_lgbm[top_features]

# Resamplear con SMOTE nuevamente
X_train_resampled, y_train_resampled = SMOTE(random_state=42).fit_resample(X_train_reduced, y_train_lgbm)
```

```python
params = {
    'objective': 'binary',
    'metric': 'auc',
    'boosting_type': 'gbdt',
    'learning_rate': 0.01,
    'num_leaves': 31,
    'max_depth': 5,
    'min_data_in_leaf': 20,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbosity': -1,
    'random_state': 42
}

train_set = lgb.Dataset(X_train_resampled, label=y_train_resampled)
valid_set = lgb.Dataset(X_valid_reduced, label=y_valid_lgbm)

model_final = lgb.train(
    params,
    train_set,
    valid_sets=[train_set, valid_set],
    num_boost_round=1000,
    callbacks=[
        lgb.early_stopping(stopping_rounds=50),
        lgb.log_evaluation(period=100)
    ]
)
```

```python
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
import lightgbm as lgb
from sklearn.metrics import roc_auc_score, classification_report

# Cargar los datos originales
df = pd.read_csv('train.csv') 


# Dividir X e y
X = df.drop(['ID_code', 'target'], axis=1)
y = df['target']

# División en train y validación
X_train, X_valid, y_train, y_valid = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Balanceo de clases con RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_train_bal, y_train_bal = ros.fit_resample(X_train, y_train)

# Verificar el nuevo balance
Counter(y_train_bal)
```

# «Para balancear el dataset altamente desbalanceado (90% clase 0 vs 10% clase 1), 
# Aplique RandomOverSampler de la librería imbalanced-learn. Esta técnica realiza un oversampling aleatorio, 
# replicando instancias existentes de la clase minoritaria hasta equiparar la cantidad de muestras con la clase mayoritaria.»

```python
# Definir los parámetros de LightGBM para un entrenamiento inicial
params = {
    'objective': 'binary',
    'metric': 'auc',
    'boosting_type': 'gbdt',
    'learning_rate': 0.01,
    'num_leaves': 31,
    'max_depth': -1,
    'verbose': -1,
    'random_state': 42
}

# Crear datasets de LightGBM
train_set = lgb.Dataset(X_train_bal, label=y_train_bal)
valid_set = lgb.Dataset(X_valid, label=y_valid)

# Entrenar el modelo con early stopping
model_lgb_balanced = lgb.train(
    params,
    train_set,
    valid_sets=[train_set, valid_set],
    num_boost_round=1000,
    callbacks=[
        lgb.early_stopping(stopping_rounds=50),
        lgb.log_evaluation(period=100)
    ]
)
```

```python
# Paso 1: Debo hacer la carga y Preprocesamiento de Datos

import pandas as pd
import numpy as np
from collections import Counter
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler

print("--- Ejecutando el Paso 1: Carga y Preprocesamiento de Datos ---")

# Cargar los datos originales
try:
    df = pd.read_csv('train.csv')
    print("Archivo 'train.csv' cargado correctamente.")
except FileNotFoundError:
    print("❌ ERROR: El archivo 'train.csv' no se encuentra. Asegúrate de que esté en el mismo directorio.")
    exit() # Salir del script si no se encuentra el archivo

# Dividir X e y
X = df.drop(['ID_code', 'target'], axis=1)
y = df['target']
print(f"Shape de X (variables): {X.shape}")
print(f"Shape de y (target): {y.shape}")

# División en train y validación
X_train, X_valid, y_train, y_valid = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
print("\nDatos divididos en conjuntos de entrenamiento y validación.")
print(f"Shape de X_train: {X_train.shape}")
print(f"Shape de X_valid: {X_valid.shape}")

# Balanceo de clases con RandomOverSampler
print("\nBalanceando el conjunto de entrenamiento...")
ros = RandomOverSampler(random_state=42)
X_train_bal, y_train_bal = ros.fit_resample(X_train, y_train)

# Verificar el nuevo balance
print("Balance de clases en el conjunto de entrenamiento original:", Counter(y_train))
print("Balance de clases en el conjunto de entrenamiento balanceado:", Counter(y_train_bal))


```

```python
# Paso 2: Entrenamiento del Modelo Random Forest 

from sklearn.ensemble import RandomForestClassifier

print("Entrenamiento del Modelo Random Forest")

# Definir los parámetros de Random Forest
params_rf = {
    'n_estimators': 300,
    'max_depth': 10,
    'random_state': 42,
    'class_weight': 'balanced',
    'n_jobs': -1
}

# Crear y entrenar el modelo Random Forest
print("Entrenando el modelo...")
model_rf = RandomForestClassifier(**params_rf)
model_rf.fit(X_train_bal, y_train_bal)

```

```python
# Paso 3: Evaluación del Modelo 

from sklearn.metrics import roc_auc_score, classification_report

print("Evaluación del Modelo Random Forest")

# Predecir probabilidades en el conjunto de validación
y_valid_pred_proba_rf = model_rf.predict_proba(X_valid)[:, 1]

# Calcular la métrica AUC
auc_score_rf = roc_auc_score(y_valid, y_valid_pred_proba_rf)
print(f"\nAUC en el conjunto de validación (Random Forest): {auc_score_rf:.4f}")

# Reporte de clasificación (usando un umbral de 0.5)
y_valid_pred_rf = (y_valid_pred_proba_rf > 0.5).astype(int)
print("\nReporte de Clasificación en el conjunto de validación (usando umbral 0.5):")
print(classification_report(y_valid, y_valid_pred_rf))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------------------
# Datos base del escenario
# ---------------------------------------
total_clientes_valid = 40000
clientes_target_1 = int(total_clientes_valid * 0.1005)  # aprox 10% clase 1 en validación

# Datos de tus modelos
modelos = {
    "LightGBM": {"recall": 0.94, "precision": 0.18},
    "RandomForest": {"recall": 0.55, "precision": 0.30}
}

# Supuestos económicos del escenario que planteare para el proyecto
costo_contacto = 5
ganancia_por_cliente = 100

# ---------------------------------------
# Calcular los datos para cada modelo
# ---------------------------------------
resultados = []

for nombre, met in modelos.items():
    recall = met["recall"]
    precision = met["precision"]

    tp = recall * clientes_target_1
    if precision > 0:
        total_contactados = tp / precision
    else:
        total_contactados = 0
    fp = total_contactados - tp

    costo_total = total_contactados * costo_contacto
    ganancia_total = tp * ganancia_por_cliente
    beneficio_neto = ganancia_total - costo_total

    resultados.append({
        "Modelo": nombre,
        "TP (captados reales)": round(tp),
        "FP (contactos sin interés)": round(fp),
        "Total contactados": round(total_contactados),
        "Costo campaña ($)": round(costo_total),
        "Ingresos ($)": round(ganancia_total),
        "Beneficio neto ($)": round(beneficio_neto)
    })

# Crear DataFrame
df_resultados = pd.DataFrame(resultados)
print("\nComparación económico-financiera de los modelos:\n")
print(df_resultados)

# ---------------------------------------
# Gráfico de barras del beneficio neto
# ---------------------------------------
plt.figure(figsize=(8,5))
plt.bar(df_resultados["Modelo"], df_resultados["Beneficio neto ($)"], color=["#4CAF50","#2196F3"])
plt.title("Beneficio neto estimado por modelo")
plt.ylabel("Beneficio neto ($)")
plt.xlabel("Modelo")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
